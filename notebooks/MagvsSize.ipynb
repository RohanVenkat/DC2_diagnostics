```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import GCRCatalogs
```

```python
from GCRCatalogs.dc2_object import DC2ObjectParquetCatalog
import astropy.units as u
```

```python
pq_path = '/global/cfs/cdirs/lsst/shared/rubin/DP0.2/objectTable/'
```

```python
gc_dp0 = DC2ObjectParquetCatalog(base_dir=pq_path, 
                                     filename_pattern='objectTable_tract_\d+\_DC2_2_2i_runs_DP0_2_v23_0_1_PREOPS-905_step3_\d+_\w+Z.parq$')
```

```python
gc_dr6 = GCRCatalogs.load_catalog('dc2_object_run2.2i_dr6')
```

```python
qtys_dp0 = ['coord_ra', 'coord_dec', 'i_free_cModelFlux', 
            'r_free_cModelFlux', 'i_free_cModelFluxErr', 'r_free_cModelFluxErr',
            'i_cModelFlux', 'r_cModelFlux', 'i_cModelFluxErr', 'r_cModelFluxErr',
            'detect_isPrimary',
            'g_free_cModelFlux','g_free_cModelFluxErr', 'g_cModelFlux','g_cModelFluxErr',
            'u_free_cModelFlux','u_free_cModelFluxErr', 'u_cModelFlux','u_cModelFluxErr',
            'z_free_cModelFlux','z_free_cModelFluxErr', 'z_cModelFlux','z_cModelFluxErr',
            'y_free_cModelFlux','y_free_cModelFluxErr', 'y_cModelFlux','y_cModelFluxErr',
            'u_ixx','u_iyy','g_ixx','g_iyy','r_ixx','r_iyy','i_ixx','i_iyy','z_ixx','z_iyy','y_ixx','y_iyy',
           ]
```

```python
qtys_dr6 = ['ra', 'dec', 'i_modelfit_CModel_instFlux', 'i_modelfit_CModel_instFluxErr', 
            'r_modelfit_CModel_instFlux', 'r_modelfit_CModel_instFluxErr', 'mag_i_cModel', 'mag_r_cModel',
           'g_modelfit_CModel_instFlux', 'g_modelfit_CModel_instFluxErr', 'mag_g_cModel',
           'u_modelfit_CModel_instFlux', 'u_modelfit_CModel_instFluxErr', 'mag_u_cModel',
           'z_modelfit_CModel_instFlux', 'z_modelfit_CModel_instFluxErr', 'mag_z_cModel',
           'y_modelfit_CModel_instFlux', 'y_modelfit_CModel_instFluxErr', 'mag_y_cModel',
            'Ixx_pixel_u','Iyy_pixel_u','Ixx_pixel_g','Iyy_pixel_g','Ixx_pixel_r','Iyy_pixel_r','Ixx_pixel_i','Iyy_pixel_i',
            'Ixx_pixel_z','Iyy_pixel_z','Ixx_pixel_y','Iyy_pixel_y',
            'cModelFlux_u','cModelFlux_g','cModelFlux_r','cModelFlux_i','cModelFlux_z','cModelFlux_y']
```

```python
data_dp0 = gc_dp0.get_quantities(qtys_dp0, native_filters='(tract > 4029) & (tract < 4033)')
```

```python
data_dr6 = gc_dr6.get_quantities(qtys_dr6, native_filters='(tract > 4029) & (tract < 4033)')
```

```python
primary = data_dp0['detect_isPrimary']
mag_i = ((data_dp0['i_cModelFlux'][primary])*u.nJy).to(u.ABmag).value
mag_r = ((data_dp0['r_cModelFlux'][primary])*u.nJy).to(u.ABmag).value
mag_g = ((data_dp0['g_cModelFlux'][primary])*u.nJy).to(u.ABmag).value
mag_u = ((data_dp0['u_cModelFlux'][primary])*u.nJy).to(u.ABmag).value
mag_z = ((data_dp0['z_cModelFlux'][primary])*u.nJy).to(u.ABmag).value
mag_y = ((data_dp0['y_cModelFlux'][primary])*u.nJy).to(u.ABmag).value

ixx_u = ((data_dp0['u_ixx'][primary])*u.nJy).to(u.ABmag).value
iyy_u = ((data_dp0['u_iyy'][primary])*u.nJy).to(u.ABmag).value
ixx_g = ((data_dp0['g_ixx'][primary])*u.nJy).to(u.ABmag).value
iyy_g = ((data_dp0['g_iyy'][primary])*u.nJy).to(u.ABmag).value
ixx_r = ((data_dp0['r_ixx'][primary])*u.nJy).to(u.ABmag).value
iyy_r = ((data_dp0['r_iyy'][primary])*u.nJy).to(u.ABmag).value
ixx_i = ((data_dp0['i_ixx'][primary])*u.nJy).to(u.ABmag).value
iyy_i = ((data_dp0['i_iyy'][primary])*u.nJy).to(u.ABmag).value
ixx_z = ((data_dp0['z_ixx'][primary])*u.nJy).to(u.ABmag).value
iyy_z = ((data_dp0['z_iyy'][primary])*u.nJy).to(u.ABmag).value
ixx_y = ((data_dp0['y_ixx'][primary])*u.nJy).to(u.ABmag).value
iyy_y = ((data_dp0['y_iyy'][primary])*u.nJy).to(u.ABmag).value
```

```python
fig, axs = plt.subplots(2,3, sharey=True, tight_layout=True,figsize=(20, 10))

axs[0][0].scatter(mag_u,ixx_u + iyy_u, label='DP0.2')
axs[0][0].set_xlabel('mag_u', fontsize=16)
axs[0][0].set_ylabel('size', fontsize=16)

axs[0][1].scatter(mag_g,ixx_g + iyy_g, label='DP0.2')
axs[0][1].set_xlabel('mag_g', fontsize=16)
axs[0][1].set_ylabel('size', fontsize=16)

axs[0][2].scatter(mag_r,ixx_r + iyy_r, label='DP0.2')
axs[0][2].set_xlabel('mag_r', fontsize=16)
axs[0][2].set_ylabel('size', fontsize=16)

axs[1][0].scatter(mag_i,ixx_i + iyy_i, label='DP0.2')
axs[1][0].set_xlabel('mag_i', fontsize=16)
axs[1][0].set_ylabel('size', fontsize=16)

axs[1][1].scatter(mag_z,ixx_z + iyy_z, label='DP0.2')
axs[1][1].set_xlabel('mag_z', fontsize=16)
axs[1][1].set_ylabel('size', fontsize=16)

axs[1][2].scatter(mag_y,ixx_y + iyy_y, label='DP0.2')
axs[1][2].set_xlabel('mag_y', fontsize=16)
axs[1][2].set_ylabel('size', fontsize=16)
```
