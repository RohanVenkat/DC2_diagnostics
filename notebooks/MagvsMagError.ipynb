```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import GCRCatalogs
import os
```

```python
from GCRCatalogs.dc2_object import DC2ObjectParquetCatalog
import astropy.units as u
```

```python
pq_path = '/global/cfs/cdirs/lsst/shared/rubin/DP0.2/objectTable/'
```

```python
gc_dp0 = DC2ObjectParquetCatalog(base_dir=pq_path, 
                                     filename_pattern='objectTable_tract_\d+\_DC2_2_2i_runs_DP0_2_v23_0_1_PREOPS-905_step3_\d+_\w+Z.parq$')
```

```python
gc_dr6 = GCRCatalogs.load_catalog('dc2_object_run2.2i_dr6')
gc_dr6.list_all_quantities()
```

```python
qtys_dp0 = ['coord_ra', 'coord_dec', 'i_free_cModelFlux', 
            'r_free_cModelFlux', 'i_free_cModelFluxErr', 'r_free_cModelFluxErr',
            'i_cModelFlux', 'r_cModelFlux', 'i_cModelFluxErr', 'r_cModelFluxErr',
            'detect_isPrimary',
            'g_free_cModelFlux','g_free_cModelFluxErr', 'g_cModelFlux','g_cModelFluxErr',
            'u_free_cModelFlux','u_free_cModelFluxErr', 'u_cModelFlux','u_cModelFluxErr',
            'z_free_cModelFlux','z_free_cModelFluxErr', 'z_cModelFlux','z_cModelFluxErr',
            'y_free_cModelFlux','y_free_cModelFluxErr', 'y_cModelFlux','y_cModelFluxErr',
           ]
```

```python
qtys_dr6 = ['ra', 'dec', 'i_modelfit_CModel_instFlux', 'i_modelfit_CModel_instFluxErr', 
            'r_modelfit_CModel_instFlux', 'r_modelfit_CModel_instFluxErr', 'mag_i_cModel', 'mag_r_cModel',
           'g_modelfit_CModel_instFlux', 'g_modelfit_CModel_instFluxErr', 'mag_g_cModel',
           'u_modelfit_CModel_instFlux', 'u_modelfit_CModel_instFluxErr', 'mag_u_cModel',
           'z_modelfit_CModel_instFlux', 'z_modelfit_CModel_instFluxErr', 'mag_z_cModel',
           'y_modelfit_CModel_instFlux', 'y_modelfit_CModel_instFluxErr', 'mag_y_cModel',
           'mag_u','mag_g','mag_r','mag_i','mag_z','mag_y',
            'magerr_u','magerr_g','magerr_r','magerr_i','magerr_z','magerr_y',
            'cModelFluxErr_u','cModelFluxErr_g','cModelFluxErr_r','cModelFluxErr_i','cModelFluxErr_z','cModelFluxErr_y',
            'magerr_y_cModel',
            'cModelFlux_u','cModelFlux_g','cModelFlux_r','cModelFlux_i','cModelFlux_z','cModelFlux_y',
           ]
```

```python
data_dp0 = gc_dp0.get_quantities(qtys_dp0, native_filters='(tract > 4029) & (tract < 4033)')
```

```python
data_dr6 = gc_dr6.get_quantities(qtys_dr6, native_filters='(tract > 4029) & (tract < 4033)')
```

```python
primary = data_dp0['detect_isPrimary']
mag_i = ((data_dp0['i_cModelFlux'][primary])*u.nJy).to(u.ABmag).value
mag_r = ((data_dp0['r_cModelFlux'][primary])*u.nJy).to(u.ABmag).value
mag_g = ((data_dp0['g_cModelFlux'][primary])*u.nJy).to(u.ABmag).value
mag_u = ((data_dp0['u_cModelFlux'][primary])*u.nJy).to(u.ABmag).value
mag_z = ((data_dp0['z_cModelFlux'][primary])*u.nJy).to(u.ABmag).value
mag_y = ((data_dp0['y_cModelFlux'][primary])*u.nJy).to(u.ABmag).value
mag_i_err = ((data_dp0['i_cModelFluxErr'][primary])*u.nJy).to(u.ABmag).value
mag_r_err = ((data_dp0['r_cModelFluxErr'][primary])*u.nJy).to(u.ABmag).value
mag_g_err = ((data_dp0['g_cModelFluxErr'][primary])*u.nJy).to(u.ABmag).value
mag_u_err = ((data_dp0['u_cModelFluxErr'][primary])*u.nJy).to(u.ABmag).value
mag_z_err = ((data_dp0['z_cModelFluxErr'][primary])*u.nJy).to(u.ABmag).value
mag_y_err = ((data_dp0['y_cModelFluxErr'][primary])*u.nJy).to(u.ABmag).value
```

```python
# print(mag_i[0], mag_i_err[0])
# print(data_dp0['i_cModelFluxErr'][primary][0])
# print(((data_dp0['i_cModelFlux'][primary][0]+data_dp0['i_cModelFluxErr'][primary][0])*u.nJy).to(u.ABmag).value - mag_i[0])
```

```python
mag_u_err_new = 2.5/np.log(10)*data_dp0['u_cModelFluxErr'][primary]/data_dp0['u_cModelFlux'][primary]
mag_g_err_new = 2.5/np.log(10)*data_dp0['g_cModelFluxErr'][primary]/data_dp0['g_cModelFlux'][primary]
mag_r_err_new = 2.5/np.log(10)*data_dp0['r_cModelFluxErr'][primary]/data_dp0['r_cModelFlux'][primary]
mag_i_err_new = 2.5/np.log(10)*data_dp0['i_cModelFluxErr'][primary]/data_dp0['i_cModelFlux'][primary]
mag_z_err_new = 2.5/np.log(10)*data_dp0['z_cModelFluxErr'][primary]/data_dp0['z_cModelFlux'][primary]
mag_y_err_new = 2.5/np.log(10)*data_dp0['y_cModelFluxErr'][primary]/data_dp0['y_cModelFlux'][primary]
```

```python
fig, axs = plt.subplots(2,3, sharey=True, tight_layout=True,figsize=(20, 10))

uds = np.arange(len(mag_u))
np.random.shuffle(uds)
mask = (mag_u[uds[:100000]]<30)
axs[0][0].scatter(mag_u[uds[:100000][mask]], mag_u_err_new[uds[:100000][mask]], label='DP0.2', marker='.',s=1, alpha=0.1)
axs[0][0].set_ylim([0,1])
axs[0][0].set_xlabel('mag_u', fontsize=16)
axs[0][0].set_ylabel('mag_u_err', fontsize=16)
axs[0][0].legend(loc='best')

gds = np.arange(len(mag_g))
np.random.shuffle(gds)
mask = (mag_g[gds[:100000]]<30)
axs[0][1].scatter(mag_g[gds[:100000][mask]], mag_g_err_new[gds[:100000][mask]], label='DP0.2', marker='.',s=1, alpha=0.1)
axs[0][1].set_ylim([0,1])
axs[0][1].set_xlabel('mag_g', fontsize=16)
axs[0][1].set_ylabel('mag_g_err', fontsize=16)
axs[0][1].legend(loc='best')

rds = np.arange(len(mag_r))
np.random.shuffle(rds)
mask = (mag_r[rds[:100000]]<30)
axs[0][2].scatter(mag_r[rds[:100000][mask]], mag_r_err_new[rds[:100000][mask]], label='DP0.2', marker='.',s=1, alpha=0.1)
axs[0][2].set_ylim([0,1])
axs[0][2].set_xlabel('mag_r', fontsize=16)
axs[0][2].set_ylabel('mag_r_err', fontsize=16)
axs[0][2].legend(loc='best')


ids = np.arange(len(mag_i))
np.random.shuffle(ids)
mask = (mag_i[ids[:100000]]<30)
axs[1][0].scatter(mag_i[ids[:100000][mask]], mag_i_err_new[ids[:100000][mask]], label='DP0.2', marker='.',s=1, alpha=0.1)
axs[1][0].set_ylim([0,1])
axs[1][0].set_xlabel('mag_i', fontsize=16)
axs[1][0].set_ylabel('mag_i_err', fontsize=16)
axs[1][0].legend(loc='best')

zds = np.arange(len(mag_z))
np.random.shuffle(zds)
mask = (mag_z[zds[:100000]]<30)
axs[1][1].scatter(mag_z[zds[:100000][mask]], mag_z_err_new[zds[:100000][mask]], label='DP0.2', marker='.',s=1, alpha=0.1)
axs[1][1].set_ylim([0,1])
axs[1][1].set_xlabel('mag_z', fontsize=16)
axs[1][1].set_ylabel('mag_z_err', fontsize=16)
axs[1][1].legend(loc='best')

yds = np.arange(len(mag_y))
np.random.shuffle(yds)
mask = (mag_y[yds[:100000]]<30)
axs[1][2].scatter(mag_y[yds[:100000][mask]], mag_y_err_new[yds[:100000][mask]], label='DP0.2', marker='.',s=1, alpha=0.1)
axs[1][2].set_ylim([0,1])
axs[1][2].set_xlabel('mag_y', fontsize=16)
axs[1][2].set_ylabel('mag_y_err', fontsize=16)
axs[1][2].legend(loc='best')
```
